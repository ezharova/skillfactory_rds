# Correlation Analysis

```python
stud_math.corr()
```

```python
stud_math.corr()[['score']].sort_values(by='score', ascending=False)
```

```python
heatmap = sns.heatmap(stud_math.corr()[['score']].sort_values(by='score', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Parameters Correlating with Students score', fontdict={'fontsize':14}, pad=10);
```

На графике корреляций видно, что самая высокая корреляция наблюдается между образованием матери и временем которое ученики тратят на учебу. При этом хотя корреляция между образованием отца и имеет значение, все же она ниже чем у матери. Это можно объяснить, что матери проводят больше времени с учениками. Высокая отрицательная корреляция наблюдается между количеством внеучебных неудач и оценкой учеников. Интересно, что возраст учеников так же влиял на успеваемость, как и время, которое ученики проводили вне дома и добирались до школы.

```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=stud_math.loc[stud_math.loc[:, column].isin(stud_math.loc[:, column].value_counts().index[:10])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
for col in ['school', 'sex', 'address', 'famsize', 'pstatus', 'mjob', 'fjob', 'reason', 'guardian', 'schoolsup',
           'famsup', 'paid', 'activities', 'nursery', 'internet', 'romantic']:
    get_boxplot(col)
```

```python
def get_stat_dif(column):
    cols = stud_math.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud_math.loc[stud_math.loc[:, column] == comb[0], 'score'], 
                        stud_math.loc[stud_math.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
for col in ['school', 'sex', 'address', 'famsize', 'pstatus', 'mjob', 'fjob', 'reason', 'guardian', 'schoolsup',
           'famsup', 'paid', 'activities', 'nursery', 'internet', 'romantic']:
    get_stat_dif(col)
```

При анализе номинативных переменных были выявлены значимые переменные Адрес учеников, работа матери и романтические отношения. Оставим в датасете эти данные для дальнейшего построния модели. 
Из первого анализа корреляций можно оставить параметры: образование матери, время которое ученик тратит на учебу, и неудачи вне школы, поскольку эти параметры показали самую высокую корреляцию. При этом я бы убрала из оценки возраст учеников, поскольку этот параметр спорный.
